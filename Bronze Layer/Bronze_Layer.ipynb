**Data** **Reading**

```python
df=spark.read.format("parquet").load("abfss://source@databricksstorageak.dfs.core.windows.net/{orders}")
df.count()
```

```python
df= spark.readStream.format("cloudFiles")\
    .option("cloudFiles.format", "parquet")\
    .option("cloudFiles.schemaLocation", f"abfss://bronze@databricksstorageak.dfs.core.windows.net/checkpoint_{filename}")\
    .load(f"abfss://source@databricksstorageak.dfs.core.windows.net/{filename}")
```

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
    .option("checkpointLocation", f"abfss://bronze@databricksstorageak.dfs.core.windows.net/checkpoint_{filename}")\
    .option("path", f"abfss://bronze@databricksstorageak.dfs.core.windows.net/{filename}")\
    .trigger(once=True)\
    .start()
```
